* Matriz Nilpotente:

Sea A una matriz de orden n, se dice que A es nilpotente de orden k si k es el menor entero positivo tal que:

$A^{k-1}=0_n$ ; $A^{k-1} [no =] a_n$

* Matriz Ortogonal:

Si el producto de la operacion es una matriz diagonal, la matriz es ortogonal.

 $A*A^T=D$

* Matriz Ortonormal:

Si el producto de la operacion es una matriz identidad, la matriz es ortonormal.

 $A*A^T=I$
