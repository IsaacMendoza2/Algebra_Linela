* Cambio de Base:

Sea u,v $\in$ V y sean $B_c$ = $[v_1,v_2,v_3,....,v_n]$ base canónoca de $\textbf{V}$ y $B'$ = $[u_1,u_2,u_3,....,u_n]$ otra base de $\textbf{V}$

Entonces: 

$[v]_{B_c}$ = $[P]^{B'}_{B_c}$ $[u]_{B'}$

$[u]_{B'}$ = $[P^{-1}]^{B'}_{B_c}$ $[v]_{B_c}$

Ejemplo:
